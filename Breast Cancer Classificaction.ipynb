### Breast Cancer Classification with a simple Neural Network (NN)

Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

```python
# loading the data to a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data,
                          columns=breast_cancer_dataset.feature_names)

data_frame.head()
```

```python
# adding the 'target' column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
y = data_frame['label']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.2,
                                                    random_state=2)
```

```python
X.shape, X_train.shape, X_test.shape
```

Building the Neural Network

```python
# importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Neural Network

history = model.fit(X_train, y_train,
                    validation_split=0.1,
                    epochs=10)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

Visualzing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'],
           loc = 'lower right')
```

Accuracy of the model on the test data

```python
loss, accuracy = model.evaluate(X_test_std, y_test)
accuracy
```

model.predict() gives the prediction probability of each class for that data point

```python
# argmax function

my_list = [0.25, 0.56]

index_of_max_value = np.argmax(my_list)
print(my_list)
print(index_of_max_value)
```

```python
# converting the prediction probability to class labels

y_pred_labels = [np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

Building the predictive system

```python
input_data = (16.84, 8.196, 51.71, 201.9, 0.086, 0.05943, 0.01588,
              0.005917, 0.1769, 0.06503, 0.1563, 0.9567, 1.094,
              8.205, 0.008968, 0.01646, 0.01588, 0.005917, 0.02574,
              0.002582, 8.964, 21.96, 57.26, 242.2, 0.1297, 0.1357,
              0.0688, 0.02564, 0.3105, 0.07409)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
prediction_label
```

```python
if (prediction_label[0] == 0):
  print('The tumor is Malignant')
else:
  print('The tumor is Benign')
```
